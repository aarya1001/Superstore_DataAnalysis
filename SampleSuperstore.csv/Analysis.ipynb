```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("SampleSuperstore.csv")
sub_category = pd.DataFrame(df.groupby("Sub-Category")["Quantity"].sum())
sub_category.reset_index(inplace=True)
sub_category = sub_category.sort_values(by="Quantity",ascending=False)
plt.bar(sub_category["Sub-Category"],sub_category["Quantity"],color="purple")
plt.xticks(sub_category["Sub-Category"],rotation='vertical')
plt.xlabel("Sub-Category")
plt.ylabel("Quantity")
plt.show()
```

WHICH STATE HAS THE HIGHEST SALES VOLUME 

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("SampleSuperstore.csv")
state_volume = pd.DataFrame(df.groupby("State")["Sales"].sum())
state_volume = state_volume.sort_values(by="Sales",ascending=False)
plt.bar(state_volume.index[0:20],state_volume["Sales"][0:20],color="Red")
plt.xticks(state_volume.index[0:20],rotation='vertical')
plt.xlabel("States")
plt.ylabel("Sales")
plt.show()
```

WHICH STATES HAS THE HIGHEST PROFIT IN SALES 

```python
state_profit = pd.DataFrame(df.groupby("State")["Profit"].sum())
state_profit = state_profit.sort_values(by="Profit",ascending=False)
plt.bar(state_profit.index[0:20],state_profit["Profit"][0:20],color="green")
plt.xticks(state_profit.index[0:20],rotation='vertical')
plt.xlabel("States")
plt.ylabel("Profit")
plt.show()
```
